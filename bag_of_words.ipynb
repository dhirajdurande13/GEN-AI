```python
# Bag OF Words IMplementation
import pandas as pd
df=pd.read_csv('./spam.csv',encoding="ISO-8859-1")
# instead of comma it will use tab seprated value for reading
```

```python
df.head()
```

```python
df.columns = df.columns.str.strip()  # Removes extra spaces or tabs
df.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], axis=1,inplace=True)  #whwn we write inplace it does'nt return anything
```

```python
df.columns = df.columns.str.strip()  # Removes spaces or tabs

df.rename(columns={'v1':'label','v2':'message'},inplace=True)
```

```python
df.head()

messages=df

messages
```

```python
# Data cleaning and preprocessing
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

ps=PorterStemmer()
```

```python
corpous=[]

for i in range(0,len(messages)):
    review=re.sub('^a-zA-Z',' ',messages['message'][i]) #inside messagges in every message apply if other than a to z or A to Z occured replace with ' 'bag_of_words.ipynb
    review=review.lower()
    # print(review) we need list
    review=review.split()
    # print(review) we got list of reviews
    # now apply steming on each word which is not stopword
    review=[ps.stem(word) for word in review if  not word in stopwords.words('English')]
    review=' '.join(review)
    # print(review)
    corpous.append(review)
```

```python
corpous
```

```python
# now apply bag of words model
# !pip install scikit-learn
from sklearn.feature_extraction.text import CountVectorizer
# cv=CountVectorizer(max_features=50)  # here focused on max_feature parameter this is Normal bag of words
cv=CountVectorizer(max_features=50,binary=True)  # here focused on max_feature parameter: this is binary BOW
```

```python
X=cv.fit_transform(corpous).toarray()
```

```python
cv.vocabulary_   #give top 50 words how many times occured this are indexes which occured most
```

```python
# for using N gram
cv=CountVectorizer(max_features=500,binary=True,ngram_range=(2,2))
# for better accuracy we can increase range of n gram
X=cv.fit_transform(corpous).toarray()
X
```

```python
messages=df
```

```python
from nltk.stem import WordNetLemmatizer

lm=WordNetLemmatizer()
```

```python
# Performing tfidf using lemmetization
corpous=[]
for i in range(0,len(messages)):
    review=re.sub('^a-zA-Z',' ',messages['message'][i])
    review=review.lower()
    review=review.split()
    review=[lm.lemmatize(word) for word in review if word not in stopwords.words('ENglish')]
    review=' '.join(review)
    corpous.append(review)
```

```python
corpous
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vc=TfidfVectorizer(max_features=100)
X=vc.fit_transform(corpous).toarray()
```
